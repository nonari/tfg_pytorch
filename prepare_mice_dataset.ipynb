```python
from glob import glob
from skimage import io
import tifffile
import matplotlib.pyplot as plt
import os
from os import path
```

```python
img_paths = glob(f'/home/nonari/Documentos/practicaempresa/santiago/VARPA/*/*/*')
```

```python
img = tifffile.imread(img_paths[0])
plt.imshow(img[:, 390:])
plt.show()
new_path = img_paths[0].replace('VARPA', 'trimmed')
os.makedirs(path.dirname(new_path), exist_ok=True)
tifffile.imwrite(new_path, img[:, 390:])
```

```python
for img_path in img_paths:
    img = tifffile.imread(img_path)
    trim = img[:, 390:]
    new_path = img_path.replace('VARPA', 'trimmed')
    os.makedirs(path.dirname(new_path), exist_ok=True)
    tifffile.imwrite(new_path, trim)
```
